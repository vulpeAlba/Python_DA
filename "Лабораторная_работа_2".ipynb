### Задание №5*

Выведите список (в виде объекта list) уникальных авторов за 2011 год, чье количество отзывовов хотя бы на одну книгу было больше 75 квантиля.

```python
df_amazon = pd.read_excel('AmazonBooks.xlsx')

df_amazon_2011 = df_amazon.query('Year == 2011')

df_amazon_quantile = float(df_amazon_2011.Reviews.quantile([0.75]))

df_amazon_2011_list = df_amazon_2011[df_amazon_2011.Reviews > df_amazon_quantile].Author

list(df_amazon_2011_list.drop_duplicates())

```

### Задание №6

Создайте столбец genre_categories содержащий данные из столбца Genre, преобразованным в тип category, а затем создайте столбец genre_codes, содержащий коды категорий.

```python
df_amazon = pd.read_excel('AmazonBooks.xlsx')
df_new_frame = pd.DataFrame(columns= ['genre_categories', 'genre_codes'])

df_new_frame.genre_categories = df_amazon.Genre.astype('category')

cat = pd.Categorical(df_amazon.Genre, ordered=True)

df_new_frame.genre_codes = df_new_frame.genre_categories.cat.codes
df_new_frame
```
